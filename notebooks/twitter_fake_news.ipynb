# Persian COVID-19 Fake News Detection — Twitter

Demonstrates the refactored pipeline using the shared `src/` package.

**Note:** LaBSE embedding downloads ~470MB on first run.

```python
import sys
from pathlib import Path

PROJECT_ROOT = Path.cwd().resolve()
if not (PROJECT_ROOT / 'src').exists():
    PROJECT_ROOT = PROJECT_ROOT.parent
sys.path.insert(0, str(PROJECT_ROOT))

from src.utils import load_config
from src.preprocessing import load_dataset
from src.training import train_all_models
from src.evaluation import summarize_training_results, metrics_table
from src.visualization import generate_all_figures

config = load_config(PROJECT_ROOT / 'config/default.yaml')
print('Config loaded.')
```

```python
dataset_path = PROJECT_ROOT / config['paths']['raw_dataset']
data = load_dataset(str(dataset_path), config)
print(f'Samples: {len(data)}')
data.head()
```

```python
results = train_all_models(data, config)
summary = summarize_training_results(results, config)
metrics_table(summary)
```

```python
results_dir = PROJECT_ROOT / 'results' / 'twitter'
metrics_df = metrics_table(summary)
generate_all_figures(results, summary, metrics_df, config, results_dir)
print(f'Figures saved to {results_dir}')
```
